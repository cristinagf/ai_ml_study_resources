```python
import torch.nn as nn
import torch

import matplotlib.pyplot as plt
torch.manual_seed(2)
```

```python
# Sigmoid functions
# Create a tensor
z = torch.arange(-10, 10, 0.1,).view(-1, 1)

# Create a sigmoid object to be used with nn.Sequential
sig = nn.Sigmoid()

#Make a prediction of the sigmoid function
y_hat = sig(z)

# Visualize results
plt.plot(z.detach().numpy(),y_hat.detach().numpy())
plt.xlabel('z')
plt.ylabel('y_hat')

# Option: Use built-in function
# y_hat = torch.sigmoid(z)
# plt.plot(z.numpy(), y_hat.numpy())
# plt.show()
```

```python
# TanH Functions
TANH = nn.Tanh()
y_hat = TANH(z)
plt.plot(z.numpy(), y_hat.numpy())
plt.show()

# Option: using the built-in tanh object
# yhat = torch.tanh(z)
# plt.plot(z.numpy(), yhat.numpy())
# plt.show()
```

```python
# ReLU function
RELU = nn.ReLU()
yhat = RELU(z)
plt.plot(z.numpy(), yhat.numpy())

# Option: Built-in function
# y_hat = torch.relu(z)
# plt.plot(z.numpy(), y_hat.numpy())
# plt.show()
```

```python
# Compare the activation functions

x = torch.arange(-2, 2, 0.1).view(-1, 1)
plt.plot(x.numpy(), torch.relu(x).numpy(), label='ReLU')
plt.plot(x.numpy(), torch.sigmoid(x).numpy(), label='Sigmoid')
plt.plot(x.numpy(), torch.tanh(x).numpy(), label='tanH')
plt.legend()
```
